```python
import glob
import os
import pandas as pd
import pickle
import shutil
```

```python
algos = [os.path.basename(x) for x in glob.glob('out/*')]
```

```python
for algo in algos:
    envs = [os.path.basename(x) for x in glob.glob('out/{}/*'.format(algo))]
    for env in envs:
        stat_files = glob.glob('out/{}/{}/*.p'.format(algo, env))
        statistics = []
        for f in sorted(stat_files):
            model_id = int(os.path.basename(f).split('_')[2].replace('.p', ''))
            run_stats = pickle.load(open(f, 'rb'))
            df = pd.DataFrame(run_stats)
            statistics.append({'reward': df['reward'].mean(), 'length': df['length'].mean(), 'model_id': model_id})
        
        result = pd.DataFrame(statistics).sort_values(by='model_id').reset_index(drop=True)
        best_result = result[result['reward'] == result['reward'].max()]
        best_model_file = 'out/{}/{}/model_round_{}.zip'.format(algo, env, best_result['model_id'].values[0])
        shutil.copy(best_model_file, 'out/{}/{}/best_model.zip'.format(algo, env))
        print('Algo: {} | Env: {}'.format(algo, env))
        display(best_result)
```
